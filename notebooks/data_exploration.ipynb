```python
import spacy

import pandas as pd
import matplotlib.pyplot as plt

from spacy import displacy
from wordcloud import WordCloud, STOPWORDS


# !python -m spacy download en_core_web_sm
nlp = spacy.load("en_core_web_sm")
```

```python
data = pd.read_csv("../data/data.csv")
data
```

```python
data["OutcomeScore"].plot.hist()
```

```python
test = nlp(data["content"][367])
displacy.render(test, style="ent")
```

```python
# spacy.explain("verb")
```

```python
wordcloud = WordCloud(width=3000, height=2000, random_state=1, background_color="salmon", colormap="Pastel1", collocations=False, stopwords=STOPWORDS).generate(data["content"].str.cat())
```

```python
plt.figure(figsize=(20, 20))
plt.imshow(wordcloud);

plt.savefig("../data/wordcloud")
```
